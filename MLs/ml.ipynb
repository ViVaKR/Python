#### Number Format
    {:d}
    {:5d}
    {:10d}.format(52)
    {:010d} # 앞에 0은 빈칸을 0으로 채움
    {:=50d} # 부호를 공백 앞에 넣기

```python
# 정수 및 부호 넣기
print("{:+d}".format(52))
print("{:=30d}".format(-52))
print("{:=+30d}".format(52))
print("{:d}".format(-100))
print("{:=+05d}".format(-52))
# 부동소수점 (전체자릿수.소숫점자릿수)
print("{:=+20.3f}".format(52.8686786)) # 자동 반올림
```

#### 숫자

- 정수 : 소수점이 없는 숫자
- 부동소수점 : 소수점이 있는 숫자
- 제곱 : **
- 나머지 : %
- 덧셈 : +
- 뺄셈 : -
- 곱셈 : *
- 나눗셈 : /
- 정수나눗셈 : // , 소수점 아래는 모두 버림
- 부동소수점 + 부동소수점 => 부동소수점
- 부동소수점 + 정수 => 부동소수점, 10 + 0.3 을 했을 때 결과로 10이라는 정수가 나오면 0.3 이라는 정보가 사라지므로 정보를 최대한 유지할 수 있게 
- 문자열 + 정수 => 불가능, 타이에러
- 문자열 + 부동소수점 => 불가능, 타입에러
- 연산자 우선순위 : * / >>> + -, 
- 제곱근(루트 연산) : 루트 4 => 4**(1/2), 3제곱근 8 => 8 ** (1/3)  

```python
print(2**9)
print(10/3)
print(4 // 2)
print(10//3)
print(10 % 3)
print(10.0 / 10.0)
print(10.0 // 3.0) # 대충 설계한 결과
print(3 + 2 * 2)    # 7, 단 괄호를 사용하는 것이 좋음
print((3 + 2) * 2)  # 10
print(8 / 2 * (2 + 2))

a = "    반갑습니다.         "
print(a.strip(), a)
```
